```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Create a pairplot to visualize relationships between variables
sns.pairplot(data, diag_kind="kde", markers="o", height=2.5)
plt.show()

# Select only the numerical columns for the correlation matrix
numerical_cols = data.select_dtypes(include=['float64', 'int64']).columns
corr_matrix = data[numerical_cols].corr()

# Create a mask to hide the upper triangle of the correlation matrix
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Set up the matplotlib figure and axis
f, ax = plt.subplots(figsize=(12, 10))

# Create the heatmap with seaborn
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', mask=mask, 
            linewidths=0.5, linecolor='white', square=True, 
            vmin=-1, vmax=1, center=0, cbar_kws={'shrink': 0.8})

# Set title and labels
plt.title('Correlation Matrix')
plt.xlabel('Features')
plt.ylabel('Features')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.show()

# Create a correlation graph with networkx
import networkx as nx
import matplotlib.pyplot as plt

# Create a correlation graph
G = nx.from_numpy_array(np.abs(corr_matrix.to_numpy()))


# Set node labels
node_labels = {i: col for i, col in enumerate(numerical_cols)}

# Draw the graph
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos, node_size=5000, node_color='lightblue', 
                 edge_color='gray', width=0.5, with_labels=False)
nx.draw_networkx_labels(G, pos, labels=node_labels, font_size=10)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style for better visualization
sns.set_style("whitegrid")

# Create a scatter plot with regression line
plt.figure(figsize=(6, 4))
sns.regplot(x="alcohol", y="quality", data=data, ci=None, line_kws={'color':'red'})

# Customize the plot
plt.title('Wine Quality vs. Alcohol Content')
plt.xlabel('Alcohol (%)')
plt.ylabel('Wine Quality (0-10)')
plt.xticks(rotation=45)
plt.grid(True)

# Add a horizontal line at the mean quality
plt.axhline(y=data['quality'].mean(), color='gray', linestyle='--', label='Mean Quality')

# Legend and layout
plt.legend(loc='upper right')
plt.tight_layout()

plt.show()
```

```python

# Create a scatter plot with regression line
plt.figure(figsize=(8, 6))
sns.regplot(x="pH", y="quality", data=data)

# Add title and labels
plt.title("pH vs. Quality")
plt.xlabel("pH")
plt.ylabel("Quality")

# Show the plot
plt.show()
```

```python
average_quality_by_sugar = data.groupby('residual sugar')['quality'].mean().reset_index()

plt.figure(figsize=(5, 3))
sns.barplot(x='residual sugar', y='quality', data=average_quality_by_sugar,hue=None, palette='coolwarm', legend=False)
plt.title('Average quality of wine by Residual sugar level')
plt.xlabel('Residual Sugar Level')
plt.ylabel('Average Quality')
plt.grid(True)
plt.show()
```
